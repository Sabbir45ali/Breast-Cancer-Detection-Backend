```python
# Importing necessary libraries

import pandas as pd 
# Pandas: Used for data manipulation and analysis
# Usage: Create, manipulate, and analyze structured data (DataFrames)
# Return Type: DataFrame, Series

import numpy as np
# Numpy: Provides support for large, multi-dimensional arrays and matrices
# Usage: Perform mathematical and logical operations on arrays
# Return Type: ndarray (N-dimensional array)

import matplotlib.pyplot as plt
# Matplotlib: Visualization library for creating static, animated, and interactive plots
# Usage: Generate plots and graphs (e.g., line, bar, scatter plots)
# Return Type: Figure, Axes (for subplots)

from sklearn.model_selection import train_test_split, GridSearchCV
# Scikit-learn modules for model training and evaluation
# Usage: Split datasets into training and testing sets
# Return Type: Tuple of arrays (e.g., (X_train, X_test, y_train, y_test))

from sklearn.preprocessing import LabelEncoder, StandardScaler
# LabelEncoder: Encodes categorical labels as integers
# Usage: Convert categorical data to numerical format
# Return Type: ndarray

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
# RandomForestClassifier & GradientBoostingClassifier: Supervised learning models for classification
# Usage: Fit classification models and make predictions
# Return Type: Fitted model object, ndarray (for predictions)

from sklearn.metrics import classification_report, accuracy_score, roc_curve, auc
# Metrics for evaluating classification performance
# Usage: Generate classification report, calculate accuracy, and ROC curve metrics
# Return Type: str (classification report), float (accuracy), ndarray (ROC metrics)

from scipy.interpolate import interp1d
# Scipy's interp1d: Interpolates a 1-D function
# Usage: Smooth ROC curves by interpolation
# Return Type: callable (interpolation function)
```

```python
# Importing the data
# Usage: Load data from a CSV file into a DataFrame for data manipulation and analysis
# Return Type: pandas.DataFrame
file_path = 'data.csv'  # Make sure the data.csv file is in the same directory
data = pd.read_csv(file_path)

# Showing dataset information
# Usage: Display concise summary of the DataFrame, including column data types and non-null values
# Return Type: None (Outputs information directly)
print("Dataset Information")
data.info()

# Displaying the first 5 rows of the dataset
# Usage: Show the top 5 rows of the DataFrame to get an overview of the data
# Return Type: pandas.DataFrame (with 5 rows)
print("\nFirst 5 Rows of the Dataset")
print(data.head())

```

```python
# Dropping unwanted columns and encoding labels

# Dropping unnecessary columns ('id' and 'Unnamed: 32')
# Usage: Remove specified columns from the DataFrame
# Return Type: pandas.DataFrame (with specified columns dropped)
data = data.drop(columns=['id', 'Unnamed: 32'])

# Encoding the 'diagnosis' column to numeric values (e.g., 0 and 1)
# Usage: Convert categorical labels to integers (e.g., Malignant to 1, Benign to 0)
# Return Type: pandas.Series (with encoded labels)
data['diagnosis'] = LabelEncoder().fit_transform(data['diagnosis'])

# Split the dataset into features (X) and target (y)

# Features (X) contain all columns except 'diagnosis'
# Usage: Separate the independent variables from the target variable
# Return Type: pandas.DataFrame
X = data.drop(columns=['diagnosis'])

# Target (y) contains only the 'diagnosis' column
# Usage: Define the dependent variable for model training
# Return Type: pandas.Series
y = data['diagnosis']

# Standardize the features

# StandardScaler: Standardizes features by removing the mean and scaling to unit variance
# Usage: Transform data to have a mean of 0 and a standard deviation of 1
# Return Type: numpy.ndarray (with scaled features)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets

# train_test_split: Splits data into training and testing subsets
# Usage: Allocate 80% of the data for training and 20% for testing
# Return Type: Tuple of numpy.ndarrays (X_train, X_test, y_train, y_test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Function to perform hyperparameter tuning and reduce overfitting

# Usage: This function takes a machine learning model and a parameter grid as input, performs hyperparameter tuning 
# using GridSearchCV, and evaluates the best model on the test set. It helps to find the optimal hyperparameters 
# and reduce overfitting.
# Return Type: Fitted model object (best_model)

def tune_and_evaluate_model(model, param_grid):
    # GridSearchCV: Performs an exhaustive search over specified hyperparameters
    # Usage: Find the best hyperparameters for the given model using cross-validation
    # Return Type: GridSearchCV object (fitted to the training data)
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    
    # Fitting the grid search model to the training data
    grid_search.fit(X_train, y_train)
    
    # Extracting the best model from grid search
    # Usage: Get the model with the best hyperparameter combination
    # Return Type: Model object (e.g., RandomForestClassifier, GradientBoostingClassifier)
    best_model = grid_search.best_estimator_
    
    # Making predictions on the test set
    # Usage: Predict target labels for the test data
    # Return Type: numpy.ndarray (predicted labels)
    y_pred = best_model.predict(X_test)
    
    # Displaying model evaluation results
    # Usage: Print classification metrics, accuracy, and best hyperparameters
    print(f"\n{type(best_model).__name__} Model Results:")
    print(classification_report(y_test, y_pred))
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Best Parameters:", grid_search.best_params_)
    
    # Returning the best model
    return best_model
```

```python
# Train a Random Forest Classifier

# Usage: Create and train a Random Forest Classifier using specified hyperparameters
# Return Type: RandomForestClassifier (fitted model)

print("\nTraining Random Forest Classifier...")

# Initializing the Random Forest Classifier with hyperparameters
rf_model = RandomForestClassifier(
    n_estimators=200,        # Number of trees in the forest (more trees can improve performance)
    max_depth=10,            # Maximum depth of each tree (controls model complexity)
    min_samples_split=5,     # Minimum samples required to split an internal node (reduces overfitting)
    min_samples_leaf=2,      # Minimum samples required to be at a leaf node (prevents overfitting)
    random_state=42          # Ensures reproducibility of results
)

# Fitting the model to the training data
# Usage: Train the model using the training dataset
# Return Type: RandomForestClassifier (fitted model)
rf_model.fit(X_train, y_train)

# Evaluate the Random Forest model

# Making predictions on the test set
# Usage: Predict target labels for the test data
# Return Type: numpy.ndarray (predicted labels)
y_pred_rf = rf_model.predict(X_test)

# Displaying classification metrics
# Usage: Generate a detailed classification report and calculate accuracy
# Return Type: str (classification report), float (accuracy score)
print("\nRandom Forest Model Results:")
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
# Tuning Random Forest Classifier

# Usage: Perform hyperparameter tuning for the Random Forest Classifier using GridSearchCV
# Return Type: Fitted RandomForestClassifier model with the best hyperparameters

print("\nTuning Random Forest Classifier...")

# Defining the parameter grid for hyperparameter tuning
# Usage: Specify a dictionary of hyperparameters with potential values to search through
# Return Type: dict (parameter grid)
rf_param_grid = {
    'n_estimators': [100, 200, 300],   # Number of trees in the forest
    'max_depth': [5, 10, 20],          # Maximum depth of each tree
    'min_samples_split': [2, 5, 10],   # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]      # Minimum samples required to be at a leaf node
}

# Tuning and evaluating the Random Forest Classifier using the custom function
# Usage: Find the best model by searching through the parameter grid
# Return Type: Fitted RandomForestClassifier model
rf_model = tune_and_evaluate_model(RandomForestClassifier(random_state=42), rf_param_grid)
```

```python
# Train a Gradient Boosting Classifier

# Usage: Create and train a Gradient Boosting Classifier using the default hyperparameters
# Return Type: GradientBoostingClassifier (fitted model)

print("\nTraining Gradient Boosting Classifier...")

# Initializing the Gradient Boosting Classifier
# Usage: Build an ensemble model that combines weak learners to improve predictive performance
# Return Type: GradientBoostingClassifier (unfitted model)
gb_model = GradientBoostingClassifier(random_state=42)

# Fitting the model to the training data
# Usage: Train the model using the training dataset
# Return Type: GradientBoostingClassifier (fitted model)
gb_model.fit(X_train, y_train)

# Evaluate the Gradient Boosting model

# Making predictions on the test set
# Usage: Predict target labels for the test data
# Return Type: numpy.ndarray (predicted labels)
y_pred_gb = gb_model.predict(X_test)

# Displaying classification metrics
# Usage: Generate a detailed classification report and calculate accuracy
# Return Type: str (classification report), float (accuracy score)
print("\nGradient Boosting Model Results:")
print(classification_report(y_test, y_pred_gb))
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
```

```python
# Model initialization with reduced overfitting settings

# Usage: Initialize machine learning models with hyperparameters that help prevent overfitting
# Return Type: RandomForestClassifier, GradientBoostingClassifier (unfitted models)

# Initializing Random Forest Classifier with reduced overfitting settings
rf_model = RandomForestClassifier(
    n_estimators=100,       # Reduced number of trees to speed up training and avoid overfitting
    max_depth=8,            # Limiting tree depth to control model complexity
    min_samples_split=5,    # Minimum samples required to split a node, preventing splits with low samples
    min_samples_leaf=2,     # Minimum samples at a leaf node, avoiding overly specific leaves
    random_state=42         # Ensuring reproducibility of results
)

# Initializing Gradient Boosting Classifier with reduced overfitting settings
gb_model = GradientBoostingClassifier(
    n_estimators=100,       # Number of boosting stages (more stages can improve performance)
    learning_rate=0.05,     # Lower learning rate with more estimators for better generalization
    max_depth=3,            # Keeping trees shallow to reduce overfitting
    random_state=42         # Consistent results across runs
)

# Training models

# Fitting the Random Forest model to the training data
# Usage: Train the Random Forest model with the training dataset
# Return Type: RandomForestClassifier (fitted model)
rf_model.fit(X_train, y_train)

# Fitting the Gradient Boosting model to the training data
# Usage: Train the Gradient Boosting model with the training dataset
# Return Type: GradientBoostingClassifier (fitted model)
gb_model.fit(X_train, y_train)
```

```python
# Function to plot ROC curve

# Usage: This function plots the Receiver Operating Characteristic (ROC) curve for a given model.
# It uses predicted probabilities to calculate the False Positive Rate (FPR) and True Positive Rate (TPR).
# The function also smooths the ROC curve using quadratic interpolation to improve visualization.
# Return Type: None (the function modifies the plot directly)

def plot_roc_curve(model, model_name, X_test, y_test):
    """
    Plots the ROC curve for a given model.
    
    Parameters:
    model (object): The trained machine learning model with a predict_proba method.
    model_name (str): The name of the model (for labeling the plot).
    X_test (numpy.ndarray or pandas.DataFrame): The test feature data.
    y_test (numpy.ndarray or pandas.Series): The true labels for the test data.
    
    Returns:
    None: The function generates a plot directly.
    """
    
    # Get predicted probabilities for the positive class (class 1)
    # Usage: Obtain probability estimates for class 1 (malignant) from the model
    # Return Type: numpy.ndarray (predicted probabilities)
    y_pred_prob = model.predict_proba(X_test)[:, 1]  
    
    # Compute FPR (False Positive Rate) and TPR (True Positive Rate)
    # Usage: Calculate FPR and TPR for different threshold values to construct the ROC curve
    # Return Type: numpy.ndarray (fpr, tpr), array (thresholds)
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    
    # Calculate the Area Under the Curve (AUC)
    # Usage: Evaluate the model's performance (higher AUC indicates better performance)
    # Return Type: float (AUC score)
    roc_auc = auc(fpr, tpr)

    # Remove duplicate FPR values to avoid interpolation issues
    # Usage: Ensure uniqueness in FPR values for smooth interpolation
    # Return Type: numpy.ndarray (unique FPR values, corresponding indices)
    unique_fpr, unique_indices = np.unique(fpr, return_index=True)
    unique_tpr = tpr[unique_indices]

    # Define a smooth curve for plotting
    # Usage: Create evenly spaced values between 0 and 1 for the smooth curve
    # Return Type: numpy.ndarray (smoothed FPR values)
    fpr_smooth = np.linspace(0, 1, 100)
    
    # Use quadratic interpolation (fallback to linear if errors persist)
    # Usage: Generate a smooth TPR curve using interpolation
    # Return Type: numpy.ndarray (smoothed TPR values)
    try:
        tpr_smooth = interp1d(unique_fpr, unique_tpr, kind='quadratic', fill_value="extrapolate")(fpr_smooth)
    except ValueError:
        tpr_smooth = interp1d(unique_fpr, unique_tpr, kind='linear')(fpr_smooth)

    # Plot the ROC curve
    # Usage: Visualize the ROC curve along with the AUC score
    # Return Type: None (adds plot to the current figure)
    plt.plot(fpr_smooth, tpr_smooth, label=f'{model_name} (AUC = {roc_auc:.2f})')
```

```python
# Plotting ROC curves for Random Forest and Gradient Boosting

# Usage: Visualize and compare the ROC curves of the Random Forest and Gradient Boosting models.
# Return Type: None (the function directly displays the plot)

# Create a new figure for the ROC curves
# Usage: Initialize a figure with specific size for better visualization
# Return Type: matplotlib.figure.Figure (created figure object)
plt.figure(figsize=(8, 6))

# Plot ROC curve for the Random Forest model
# Usage: Generate and add the ROC curve of Random Forest to the plot
# Return Type: None (the function modifies the plot directly)
plot_roc_curve(rf_model, "Random Forest", X_test, y_test)

# Plot ROC curve for the Gradient Boosting model
# Usage: Generate and add the ROC curve of Gradient Boosting to the plot
# Return Type: None (the function modifies the plot directly)
plot_roc_curve(gb_model, "Gradient Boosting", X_test, y_test)

# Customize the plot with additional elements

# Add a diagonal reference line indicating a random classifier's performance
# Usage: Draw a dashed line from (0, 0) to (1, 1) as a reference
# Return Type: list (line objects created)
plt.plot([0, 1], [0, 1], 'k--')

# Add labels to the x and y axes
# Usage: Provide meaningful labels for better plot interpretation
# Return Type: Text (matplotlib text object)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

# Add a title to the plot
# Usage: Set the plot title for clarity
# Return Type: Text (matplotlib text object)
plt.title("ROC Curves for Random Forest and Gradient Boosting")

# Add a legend to the plot to differentiate between models
# Usage: Place the legend in the lower right corner
# Return Type: Legend (matplotlib legend object)
plt.legend(loc="lower right")

# Add a grid to the plot for easier interpretation
# Usage: Enable grid lines on the plot
# Return Type: None (modifies the plot directly)
plt.grid()

# Display the final plot
# Usage: Render the plot in the output
# Return Type: None (the plot is shown directly)
plt.show()
```

```python
# Function to take manual input and make a prediction

# Usage: This function allows users to input feature values manually through the console.
# It scales the input data, uses the trained model to predict the class (Malignant or Benign),
# and displays the result in a user-friendly format.
# Return Type: None (the function directly outputs the prediction result to the console)

def manual_input_prediction(model, scaler):
    """
    Takes manual input for all features, scales the input, and predicts the result using the model.
    
    Parameters:
    model (object): The trained machine learning model (must support the predict method).
    scaler (object): The scaler used for feature standardization (e.g., StandardScaler).
    
    Returns:
    None: Displays the prediction result in the console.
    """
    
    # Getting input from the user for each feature
    print("Enter the following feature values:")
    
    # Feature names from the dataset
    # Usage: Retrieve the feature names from the original dataset
    # Return Type: pandas.Index (column names of the dataset)
    feature_names = X.columns
    
    # List to store input values
    # Usage: Accumulate user inputs in the correct order
    # Return Type: list (containing feature values as floats)
    input_data = []
    
    # Loop through each feature and collect input
    # Usage: Prompt the user for input for each feature
    # Return Type: None (appends values to input_data list)
    for feature in feature_names:
        try:
            value = float(input(f"{feature}: "))  # Convert input to float for model compatibility
            input_data.append(value)
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
            return
    
    # Convert input data to a DataFrame
    # Usage: Prepare the input data for scaling and prediction
    # Return Type: pandas.DataFrame (single row with feature values)
    input_df = pd.DataFrame([input_data], columns=feature_names)
    
    # Scale the input using the same scaler used during model training
    # Usage: Standardize the input data to match the model's expectations
    # Return Type: numpy.ndarray (scaled feature values)
    input_scaled = scaler.transform(input_df)
    
    # Make the prediction using the model
    # Usage: Predict the class (Malignant or Benign) based on the input data
    # Return Type: numpy.ndarray (predicted class as an integer array)
    prediction = model.predict(input_scaled)
    
    # Interpret and display the result
    # Usage: Provide a human-readable prediction result
    # Return Type: None (prints the result directly)
    result = 'Malignant' if prediction[0] == 1 else 'Benign'
    print("\nPrediction Result:", result)


# Using the Random Forest model for manual input prediction
# Usage: Call the function with the trained model and scaler
# Return Type: None (prompts user for input and displays the result)
manual_input_prediction(rf_model, scaler)
```
